```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Derivative Calculator Function with backward, forward & central option
def my_der_calc(f,a,b,n,option):
  x=np.linspace(a,b,n)
  if option == "forward":
    X=x[0:n-1]
    df=(f(x[1:n])-f(x[0:n-1]))/(x[1:n]-x[0:n-1])
  elif option == "backward":
    X=x[1:n]
    df=(f(x[1:n])-f(x[0:n-1]))/(x[1:n]-x[0:n-1])
  else :
    X=x[1:n-1]
    df=(f(x[2:n])-f(x[0:n-2]))/(x[2:n]-x[0:n-2])
  return df,X

def plot_graph(f, message):
  plt.figure(figsize=(12,8))
  plt.plot(x,np.cos(x),label="analytic")
  [df,X]=my_der_calc(f,0,4*np.pi,10,"forward")
  plt.plot(X,df,label="forward")
  [df,X]=my_der_calc(f,0,4*np.pi,10,"backward")
  plt.plot(X,df,label="backward")
  [df,X]=my_der_calc(f,0,4*np.pi,10,"central")
  plt.plot(X,df,label="central")
  plt.legend()
  plt.title(message)
  plt.xlabel("X")
  plt.ylabel("Y")
  plt.show()
```

```python
x=np.linspace(0,4*np.pi,100)
f=lambda x:np.sin(x)
plot_graph(f, "Analytic and numerical derivatives of Sin")
```

```python
f=lambda x:np.cos(x)
plot_graph(f, "Analytic and numerical derivatives of Cos")
```

```python
f=lambda x:np.tan(x)
plot_graph(f, "Analytic and numerical derivatives of Tan")
```
