```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-poster')
```

```python
# Linear Interpolation function 
# Parameters x and y points alongside list of X 
# X[i] points which will get value of Y^[i]

def my_lin_interp(x, y, X):
  Y = []
  for i in range(0, len(X)):
    _i = int(X[i])
    output = y[_i] + (X[i] - x[_i]) * ((y[_i+1] - y[_i])/(x[_i+1] - x[_i]))
    Y.append(output)
  return Y
```

```python
# Testing linear interpolation function
x = [0, 1, 2]
y = [1, 3, 2]

print('------------------------')
print(my_lin_interp(x, y, [1.5])) # Should print [2.5]

print('------------------------')
print(my_lin_interp(x, y, [0.5, 1.0, 1.5])) # Should print [2.0, 3.0, 2.5]

print('------------------------')
# Now validating our result using scipy interpolation
f = interp1d(x, y)
# For 0.5 y^hat should print 2.0
print('For 0.5 y^hat should print 2.0')
y_hat = f(0.5)
print(y_hat)

print('------------------------')
# For 1.0 y^hat should print 3.0
print('For 1.0 y^hat should print 3.0')
y_hat = f(1.0)
print(y_hat)

print('------------------------')
# For 1.5 y^hat should print 2.5
print('For 1.5 y^hat should print 2.5')
y_hat = f(1.5)
print(y_hat)
```

```python
# Getting y_hat from our local lin_interp function
y_hat = my_lin_interp(x, y, [0.5, 1.0, 1.5])
                      
plt.plot(x, y, '-ob')
plt.plot(0.5, y_hat[0], 'ro')
plt.title('Linear Interpolation at x = 0.5')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
