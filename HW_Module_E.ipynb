```python
from sympy import *

# Derivative Calculation

def derivative(f, x):
  return f.diff(x)

x = Symbol('x')
y = x**2 + 1
print(f'Derivative of x^2 + 1 =  {derivative(y, x)}') # 2x

t = derivative(y, x)

y = x**3 + x**2 + x
print(f'Derivative of x^3 + x^2 + x =  {derivative(y, x)}') # 3x^2 + 2x + 1

y = 5*x 
print(f'Derivative of 5x =  {derivative(y, x)}') # 5
```

```python
import matplotlib.pyplot as plt
import numpy as np

def function(val):
  return val**2 + 1

def derivative(val):
  return 2*val

y = np.linspace(-12, 12)

plt.plot(y, function(y), color='purple', label='Function x^2 + 1')
  
# plotting its derivative
plt.plot(y, derivative(y), color='green', label='Derivative 2x')
  
# formatting
plt.legend(loc='upper left')
plt.grid(True)
```

```python
def function1(val):
  return val**3 + val**2 + val

def derivative1(val):
  return 3*val**2 + 2*val + 1

y = np.linspace(-12, 12)

plt.plot(y, function1(y), color='purple', label='Function x^3 + x^2 + x')
  
# plotting its derivative
plt.plot(y, derivative1(y), color='green', label='Derivative 3x^2 + 2x + 1')
  
# formatting
plt.legend(loc='upper left')
plt.grid(True)
```

```python
# Limit Calculation
x = Symbol('x')
y = sin(x)/x

# when lim approaches 0
print(f'sin(x)/x at x->0 = {limit(y, x, 0)}') ## 1

# when lim approaches oo
print(f'sin(x)/x at x->oo = {limit(y, x, oo)}') ## 0


print()

y = cos(x)/x

# when lim approaches 0
print(f'cos(x)/x at x->0 = {limit(y, x, 0)}') ## oo Infinity

# when lim approaches oo
print(f'cos(x)/x at x->oo = {limit(y, x, oo)}') ## 0

print()

y = tan(x)/x

# when lim approaches 0
print(f'tan(x)/x at x->0 = {limit(y, x, 0)}') ## 1
```

```python
import math

# Taylor Series & Error Calculation on function e^x

def func_e(x, n):
    e_approx = 0
    for i in range(n):
        e_approx += x**i/math.factorial(i)    
    return e_approx

x = 5
for i in range(1,11):
    e_approx = func_e(x,i)
    e_exp = math.exp(x)
    e_error = abs(e_approx - e_exp)
    print(f'{i} terms: Taylor Series approx= {e_approx}, exp calc= {e_exp}, error = {e_error}')
```
